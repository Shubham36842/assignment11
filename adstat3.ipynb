```python
import numpy as np
def population_mean(sm,sd,n):
    val=sm+1.96*(sd/np.sqrt(n))
    return val
```

```python
population_mean(5,4,10)
```

```python
# Python program to demonstrate how to
# perform two sample T-test

# Import the library
import scipy.stats as stats

# Creating data groups
data_group1 = np.array([14, 15, 15, 16, 13, 8, 14,
						17, 16, 14, 19, 20, 21, 15,
						15, 16, 16, 13, 14, 12])

data_group2 = np.array([15, 17, 14, 17, 14, 8, 12,
						19, 19, 14, 17, 22, 24, 16,
						13, 16, 13, 18, 15, 13])

# Null Hypothesis-No difference between population mean
#Alternate Hypothesis- difference between population mean
statics,pval=stats.ttest_ind(a=data_group1, b=data_group2, equal_var=True)
if(pval>0.05):
    print("we fail to reject the null hypothesis")
else:
    print("we reject the null hypothesis")
```

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import t

x = np.linspace(-5, 5, 100)
y = t.pdf(x, 10)
plt.plot(x, y, label=f"Degrees of Freedom = 10")

plt.xlabel('x')
plt.ylabel('PDF')
plt.title('T-Distribution with  Degrees of Freedom=10')
plt.legend()
plt.show()
```

```python
# Python program to demonstrate how to
# perform two sample T-test

# Import the library
import scipy.stats as stats

# Creating data groups
data_group1 = np.array([14, 15, 15, 16, 13, 8, 14,
						17, 16, 14, 19, 20, 21, 15,
						15, 16, 16, 13, 14, 12])

data_group2 = np.array([15, 17, 14, 17, 14, 8, 12,
						19, 19, 14, 17, 22, 24, 16,
						13, 16, 13, 18, 15, 13])

# Perform the two sample t-test with equal variances
#Null Hypothesis- mean of sample 1 equal to mean of sample 2
#Alternate Hypothesis- mean of sample 1 is not equal to sample 2
statics,pval=stats.ttest_ind(a=data_group1, b=data_group2, equal_var=True)
if(pval>0.05):
    print("we fail to reject the null hypothesis")
else:
    print("we reject the null hypothesis")
```
